```python
income=income[['GEO.id','Household_total','Median_income']]
```

```python
income = income.rename(columns={'GEO.id':'Id'})
```

```python
park= pd.read_csv('/Users/shiyuan/Documents/Machine learning/project-park crime/data/10 park_census_tract.csv')
```

```python
del park['Unnamed: 0']
```

```python
park
```

```python
park['ct']=map(lambda x: x.split(',')[:-1], park['census_tract'])
```

```python
park['household']=0
park['median_income']=0
```

```python
park.head(10)
```

```python
import numpy as np
```

```python

for i in range(len(park)):
    for j in park.ix[i,'ct']:
        park.ix[i,'household'] += income[income['Id']==j]['Household_total'].values[0]
        park.ix[i,'median_income']=np.mean(income[income['Id'].isin(park.ix[i,'ct'])]['Median_income'])
```
